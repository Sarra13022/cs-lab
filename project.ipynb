```python
#split dataset in features and target variable
X = dff.drop(['annual_consume'], axis='columns')
y = dff['annual_consume']
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
# Using GridSearchCV to find the best algorithm for this problem
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
```

```python
def find_best_model(X, y):
    models = {
        'linear_regression': {
            'model': LinearRegression(),
            'parameters': {
                'normalize': [True,False]
            }
        },
        
        'lasso': {
            'model': Lasso(),
            'parameters': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'parameters': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    
    scores = []
    cv_X_y = ShuffleSplit(n_splits=5, test_size=0.20, random_state=0)
    
    for model_name, model_params in models.items():
        gs = GridSearchCV(model_params['model'], model_params['parameters'], cv=cv_X_y, return_train_score=False)
        gs.fit(X_train, y_train)
        y_pred=gs.predict(X_test)
        y_pred_example=gs.predict(X_test.loc[10,:].to_numpy().reshape(1,-1))
                                  
        Rmse=np.sqrt(metrics.mean_squared_error(y_test ,y_pred))
        scores.append({
            'model': model_name,
            'best_parameters': gs.best_params_,
            'R_squared': gs.best_score_, 
            'RMSE': Rmse,
            'example_annual_consumption': y_test.loc[10],
            'example_predicted_annual_consumption': y_pred_example
       
        })
        
    return pd.DataFrame(scores, columns=['model', 'best_parameters', 'R_squared','RMSE','example_annual_consumption','example_predicted_annual_consumption'])
find_best_model(X, y)
```

```python
dff['annual_binary'] = np.where(dff['annual_consume']<2330660, 0, 1)    
dff
```

```python
# import the class
from sklearn.linear_model import LogisticRegression
X= dff.drop(['annual_consume','annual_binary'], axis='columns')
y = dff['annual_binary']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
 #instantiate the model (using the default parameters)
logreg = LogisticRegression(solver='lbfgs', max_iter=100)
# fit the model with data
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual consumption')
plt.xlabel('Predicted consumption')
```
